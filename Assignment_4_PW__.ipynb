### 4) What is the difference between abstraction and encapsulation?

**Answer:**

Abstraction: Hides complex implementation details and exposes only the essential interface.

Encapsulation: Bundles data and methods together and controls access to internal state (information hiding).

Example (both together):

```
class CoffeeMachine:
    def __init__(self):
        self._water_ml = 200   # encapsulated-ish 

    # ABSTRACTION: user just presses a button; internals are hidden
    def make_espresso(self):
        if self._water_ml < 50:
            return "Add water, barista!"
        self._water_ml -= 50
        return "â˜• Espresso ready!"

    # ENCAPSULATION: controlled access via a property
    @property
    def water_ml(self):
        return self._water_ml

    @water_ml.setter
    def water_ml(self, ml):
        if ml < 0:
            raise ValueError("No negative water, we're not making dust coffee.")
        self._water_ml = ml

```
